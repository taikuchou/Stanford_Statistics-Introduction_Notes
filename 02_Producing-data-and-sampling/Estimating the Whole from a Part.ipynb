## Statistics: Estimating the Whole from a Part

### Key Terms

- **Population**: the **entire group** weâ€™re interested in (e.g., all residents of Canada).

- **Parameter**: the true value for the population (e.g., the actual average age of all Canadians).

- **Sample**: a **smaller group** selected from the population (e.g., 1,000 Canadians surveyed).

- **Statistic/Estimate**: the **calculated value** from the sample, used to approximate the population parameter (e.g., the average age from the survey).

**Example**: Instead of asking every single Canadian their average income (impossible in practice), we survey a random sample. The sample average is our statistic, which estimates the true population average (parameter).

When we collect data, how we choose the sample matters. If the sample doesnâ€™t represent the population fairly, our results will be biased and misleading. Two main methods to reduce bias are:
- simple random sampling
- stratified random sampling.

#### Bias in Sampling

- **Convenience sampling** (e.g., asking only voters in your hometown) is biased because it misses diversity in the whole population.

- **Common biases**:

    - **Selection bias** â€“ some groups are more likely to be included.

    - **Non-response bias** â€“ certain people donâ€™t respond (e.g., busy individuals).

    - **Voluntary response bias** â€“ only highly motivated people respond (e.g., online polls).

Bias skews results away from the true population picture.

### Methods to Avoid Bias
1. **Simple Random Sampling**

- Each individual in the population has an **equal chance** of being chosen.

- Like pulling names from a hat.

- Advantage: **Unbiased, fair representation.**

- Limitation: May not capture differences between subgroups if they exist.

2. **Stratified Random Sampling**

- The population is divided into **strata (subgroups)**, such as age, gender, or income level.

- A random sample is taken **within each subgroup**.

- Advantage: Often **more precise**, ensures subgroups are represented.

- Limitation: **More complex** and requires knowledge of the population structure.

To understand cause-and-effect relationships, we need to distinguish between **observations** and **experiments**, be aware of **confounding factors**, and account for the **placebo effect.**

### Observation vs. Experiment

- **Observational Study**

    - Researchers only **observe** what people naturally do.

    - **Example**: Tracking peopleâ€™s red meat consumption and cancer rates.

    - **Problem**: Cannot prove causation because **confounding variables** (e.g., lifestyle, genetics) may explain the association.

- **Experiment**

    - Researchers **assign treatments** to groups (treatment vs. control).

    - Random assignment helps ensure groups are similar except for the treatment.

    - **Advantage**: Can establish causal relationships.

### Confounding Variables

- A **confounding variable** is a hidden factor that influences both the supposed cause and effect.

- **Example**: People who eat more red meat may also exercise less â†’ Is cancer risk from meat, or from less exercise?

ðŸ‘‰ Confounding makes it impossible to prove cause from observation alone.

### The Placebo Effect

- A **placebo** is a fake treatment (e.g., sugar pill) given to the control group.

- The **placebo effect**: People improve simply because they **believe** they are receiving treatment.

- This effect is both psychological and biological (e.g., stress reduction, endorphin release).

### Double-Blind Experiments

- **Single-blind**: Participants donâ€™t know if they got real treatment or placebo.

- **Double-blind**: Neither participants nor evaluators know group assignments.

Purpose: Prevent bias from expectations of both patients and researchers.

![Observation VS Experiment](./images/Observation_Experiment.png)

**Randomized controlled experiments (RCTs)** are designed to establish causal relationships by eliminating bias from confounders and using chance to make comparisons fair.

ðŸ”¹ **Why Random Assignment Matters**

- **Treatment vs. Control Groups**: Subjects are split into groups. One receives the treatment, the other serves as a baseline.

- **Random Assignment**: Each subject is randomly assigned, so known and unknown confounders (e.g., age, lifestyle, genetics) are, on average, balanced across groups.

- **Result**: Differences in outcomes can be more confidently attributed to the treatment, not to preexisting differences.

ðŸ”¹ **Chance Effects**

- Even with randomization, **groups may differ by chance** (e.g., one group might get slightly more older people).

- The beauty of randomization is that the **size of these chance differences can be calculated** statistically (through probability models).

- Larger sample sizes make these chance imbalances smaller and easier to quantify.

ðŸ”¹ **Connection to Sampling and Error**

- Data come from a sample of the population, used to estimate parameters.

- **Statistic = Parameter + Bias + Chance Error**

    - **Bias comes** from poor design (not random).

    - **Chance error** comes from randomness but can be measured.

- **Random sampling** (for surveys) and **random assignment** (for experiments) both use chance to control bias and quantify error.

ðŸ”¹ **Place in the Bigger Picture**

- **Observational studies**: show associations but suffer from confounding.

- **Randomized controlled experiments**: balance confounders, control placebo effects, and quantify chance error â†’ strongest evidence for causation.

- **Statistics overall**: give us tools to measure and control error, so we can make reliable inferences about large populations using only manageable samples.

### Takeaway

- **Random assignment** â†’ balances confounders across groups.

- **Chance error** â†’ unavoidable but quantifiable, and decreases with larger samples.

- **Bias** â†’ systematic error that must be avoided through good design.

- **RCTs** â†’ the gold standard for determining causal effects in science and medicine.

### 1. Simple Random Sampling (SRS) vs Stratified Random Sampling

```python
import numpy as np
```

```python
rng = np.random.default_rng(0)
rng,type(rng)
```

```python
# --- Build a population with 3 strata having different means ---
N = 120_000
strata_props = np.array([0.2, 0.5, 0.3])
strata_mu    = np.array([50,   60,   70])   # different subgroup means
strata_sd    = np.array([10,   12,   11])

strata_sizes = (strata_props * N).astype(int)
labels = np.repeat(np.arange(3), strata_sizes)
```

```python
strata_sizes, labels
```

```python
# population outcome Y
Y = rng.normal(loc=strata_mu[labels], scale=strata_sd[labels])
pop_mean = Y.mean()
print(f"Population mean = {pop_mean:.2f}")
```

```python
# --- SRS: draw n at random from the whole population ---
def srs_mean(n=1000):
    idx = rng.choice(N, size=n, replace=False)
    return Y[idx].mean()

# --- Stratified: sample proportionally from each stratum ---
def stratified_mean(n=1000):
    out = []
    for g in range(3):
        idx_g = np.where(labels == g)[0]
        n_g = int(round(n * strata_props[g]))  # proportional allocation
        out.append(rng.choice(idx_g, size=n_g, replace=False))
    idx = np.concatenate(out)
    return Y[idx].mean()

# Compare sampling variability by Monte Carlo
def mc_var(estimator, reps=2000, n=1000):
    ests = np.array([estimator(n) for _ in range(reps)])
    return ests.mean(), ests.std()
```

```python
for n in [200, 500, 1000, 2000]:
    m_srs, sd_srs   = mc_var(srs_mean, n=n)
    m_str, sd_strat = mc_var(stratified_mean, n=n)
    print(f"n={n:>4} | SRS: biasâ‰ˆ{m_srs-pop_mean:>7.3f}, SD={sd_srs:>6.3f} | "
          f"Strat: biasâ‰ˆ{m_str-pop_mean:>7.3f}, SD={sd_strat:>6.3f}")
```

**What to observe**: Both are (approximately) unbiased, but **stratified sampling typically has a smaller SD (higher precision)** when strata differ meaningfully

## 2. Sampling Bias vs Chance Error (estimate = parameter + bias + chance)

```python
rng = np.random.default_rng(1)
```

```python
# Population with location effect (urban vs rural) + noise
N = 200_000
is_urban = rng.random(N) < 0.6
true_effect = 0.0                    # no true global difference we care about here
base = 100 + 10*is_urban + rng.normal(0, 8, N)  # urban avg is 10 higher

parameter = base.mean()
print(f"True parameter (population mean): {parameter:.2f}")
```

```python
def srs(n=1000):
    idx = rng.choice(N, size=n, replace=False)
    return base[idx].mean()

def convenience(n=1000):  # biased: sample mostly urban
    idx_urban = np.where(is_urban)[0]
    idx = rng.choice(idx_urban, size=n, replace=False)  # ignores rural
    return base[idx].mean()

# Decompose estimate - parameter across many reps
def decompose(estimator, reps=2000, n=1000):
    ests = np.array([estimator(n) for _ in range(reps)])
    bias = ests.mean() - parameter
    chance_sd = ests.std()           # SD â‰ˆ size of chance error
    return bias, chance_sd
```

```python
for n in [200, 500, 1000, 5000]:
    b_srs, se_srs = decompose(srs, n=n)
    b_con, se_con = decompose(convenience, n=n)
    print(f"n={n:>4} | SRS:    bias={b_srs:>7.3f}, chanceâ‰ˆ{se_srs:>6.3f}")
    print(f"      | Conv.:  bias={b_con:>7.3f}, chanceâ‰ˆ{se_con:>6.3f}")
```

### What to observe:

- **Chance error shrinks with larger n** (SRS rows).

- **Bias remains** even with large n (convenience sampling rows).

## 3. Standard Error and Confidence Interval for a Mean

```python
import numpy as np
from math import sqrt
rng = np.random.default_rng(2)
```

```python
# Unknown population; we'll pretend we only have sample data
pop = rng.normal(loc=65, scale=15, size=500_000)
true_mu = pop.mean()
```

```python
def mean_ci(sample, alpha=0.05):
    xbar = sample.mean()
    s = sample.std(ddof=1)
    n = sample.size
    se = s / sqrt(n)
    from scipy.stats import t
    tcrit = t.ppf(1 - alpha/2, df=n-1)
    lo, hi = xbar - tcrit*se, xbar + tcrit*se
    return xbar, se, (lo, hi)
```

```python
for n in [30, 100, 400, 1600]:
    samp = rng.choice(pop, size=n, replace=False)
    xbar, se, (lo, hi) = mean_ci(samp)
    covered = (lo <= true_mu <= hi)
    print(f"n={n:>4} | xÌ„={xbar:6.2f}  SE={se:5.2f}  CI=({lo:6.2f},{hi:6.2f})  covers? {covered}")
```

What to observe: SE âˆ 1/âˆšn, and CIs get tighter as n grows.

## 4. Observational vs Randomized (Confounding Demo)

```python
rng = np.random.default_rng(3)

N = 100_000
# Confounder: health-consciousness H ~ N(0,1)
H = rng.normal(0, 1, N)
# Treatment assignment in an observational setting depends on H
p_treat_obs = 1 / (1 + np.exp(-1.2*(H - 0.3)))   # more health-conscious -> more likely treatment
T_obs = rng.binomial(1, p_treat_obs)
```

```python
# True causal effect of treatment on outcome Y
tau = 2.0

# Outcome depends on treatment and H (confounding), plus noise
epsilon = rng.normal(0, 1.5, N)
Y = 50 + tau*T_obs + 3*H + epsilon

# Observational estimate: difference in means (biased by H-imbalance)
obs_est = Y[T_obs==1].mean() - Y[T_obs==0].mean()

# Randomized experiment: assign T at random (balances H on average)
T_rct = rng.binomial(1, 0.5, N)
Y_rct = 50 + tau*T_rct + 3*H + rng.normal(0, 1.5, N)
rct_est = Y_rct[T_rct==1].mean() - Y_rct[T_rct==0].mean()

print(f"True effect tau        = {tau:.2f}")
print(f"Observational estimate = {obs_est:.2f}  (biased by confounding)")
print(f"RCT estimate           = {rct_est:.2f}  (approximately unbiased)")
```

### What to observe: 
The **observational estimate** is pulled away from the true effect by confounding. The **RCT estimate** is close to the truth.

## 5. Placebo Effect & Double-Blind Measurement Bias (Simulation)

```python
rng = np.random.default_rng(4)

N = 10_000
true_tau = 1.0  # real physiological treatment effect
placebo_prop = 0.25   # fraction of people who respond to placebo belief
```

```python
# Randomized assignment
T = rng.binomial(1, 0.5, N)

# Placebo responders in control group (belief-induced improvement)
placebo_responder = rng.random(N) < placebo_prop
belief_gain = np.where((T==0) & placebo_responder, 0.8, 0.0)

# True outcome (biology)
baseline = rng.normal(50, 8, N)
Y_true = baseline + true_tau*T

# Measurement bias if evaluator is NOT blinded (expects treatment to work)
evaluator_bias = np.where(T==1, 0.5, -0.1)  # subtle over/under scoring

# Observed outcomes under single-blind (participants blinded, evaluator not)
Y_single_blind = Y_true + belief_gain + evaluator_bias + rng.normal(0, 1.2, N)

# Observed outcomes under double-blind (no evaluator bias)
Y_double_blind = Y_true + belief_gain + rng.normal(0, 1.2, N)

def diff_in_means(Y, T):
    return Y[T==1].mean() - Y[T==0].mean()

print(f"True effect (biology only)        : {true_tau:.2f}")
print(f"Estimated effect (single-blind)   : {diff_in_means(Y_single_blind, T):.2f}  <-- inflated by evaluator bias")
print(f"Estimated effect (double-blind)   : {diff_in_means(Y_double_blind, T):.2f}  <-- closer to truth")
```

### What to observe:

- Placebo responders lift control outcomes.

- **Single-blind** (unblinded evaluator) **inflates** the estimate.

- **Double-blind** removes evaluator bias, yielding a more accurate effect.
