# Correlation & Regression â€“ Jupyter Notebook

This notebook demonstrates correlation, regression, prediction, regression to the mean, and normal approximation given x.

## 1. Setup & Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

plt.rcParams["figure.figsize"] = (6, 4)
plt.rcParams["axes.grid"] = True

np.random.seed(42)
```

## 2. Simulate Midterm & Final Exam Data with Given Correlation

```python
n = 300
mean_midterm = 50
mean_final = 50
sd_midterm = 10
sd_final = 10
r = 0.6

cov = [
    [sd_midterm**2, r * sd_midterm * sd_final],
    [r * sd_midterm * sd_final, sd_final**2],
]

midterm, final = np.random.multivariate_normal(
    mean=[mean_midterm, mean_final], cov=cov, size=n
).T

df = pd.DataFrame({"midterm": midterm, "final": final})
df.head()
```

## 3. Scatterplot & Correlation

```python
plt.scatter(df["midterm"], df["final"], alpha=0.5)
plt.xlabel("Midterm score")
plt.ylabel("Final score")
plt.title("Midterm vs Final")
plt.show()

r_empirical = df["midterm"].corr(df["final"])
r_empirical
```

## 4. Regression Line by Formula

```python
x = df["midterm"].values
y = df["final"].values

x_mean = x.mean()
y_mean = y.mean()
sx = x.std(ddof=1)
sy = y.std(ddof=1)

b = r_empirical * (sy / sx)
a = y_mean - b * x_mean

a, b
```

## 5. Plot Regression Line

```python
x_grid = np.linspace(x.min(), x.max(), 100)
y_hat_grid = a + b * x_grid

plt.scatter(x, y, alpha=0.4, label="Data")
plt.plot(x_grid, y_hat_grid, linewidth=2, label="Regression line")
plt.xlabel("Midterm")
plt.ylabel("Final")
plt.legend()
plt.show()
```

## 6. Predict Final from Midterm

```python
def predict_final_from_midterm(midterm_score):
    return a + b * midterm_score

for score in [30, 50, 70, 90]:
    print(f"Midterm = {score} â†’ Predicted final = {predict_final_from_midterm(score):.2f}")
```

## 7. Normal Approximation Given x

```python
def conditional_sd_y_given_x(r, sy):
    return np.sqrt(1 - r**2) * sy

cond_sd = conditional_sd_y_given_x(r_empirical, sy)
cond_sd
```

```python
def prob_final_above_cutoff_given_midterm(cutoff, midterm_score):
    mu_y_given_x = predict_final_from_midterm(midterm_score)
    sd_y_given_x = conditional_sd_y_given_x(r_empirical, sy)
    return stats.norm.sf(cutoff, loc=mu_y_given_x, scale=sd_y_given_x)

prob_final_above_cutoff_given_midterm(60, 40)
```

## 8. Regression to the Mean

```python
high_threshold = x_mean + 2 * sx
low_threshold = x_mean - 2 * sx

high_group = df[df["midterm"] > high_threshold]
low_group = df[df["midterm"] < low_threshold]

print("High midterm group mean midterm:", high_group["midterm"].mean())
print("High midterm group mean final:", high_group["final"].mean())

print("Low midterm group mean midterm:", low_group["midterm"].mean())
print("Low midterm group mean final:", low_group["final"].mean())
```

```python
plt.scatter(df["midterm"], df["final"], alpha=0.3, label="All students")
plt.scatter(high_group["midterm"], high_group["final"], color="red", label="High midterm")
plt.scatter(low_group["midterm"], low_group["final"], color="green", label="Low midterm")
plt.xlabel("Midterm")
plt.ylabel("Final")
plt.legend()
plt.show()
```

## 9. Predict Midterm from Final (Different Regression Line)

```python
x2 = df["final"].values
y2 = df["midterm"].values

x2_mean = x2.mean()
y2_mean = y2.mean()
sx2 = x2.std(ddof=1)
sy2 = y2.std(ddof=1)

b2 = r_empirical * (sy2 / sx2)
a2 = y2_mean - b2 * x2_mean

print(f"Final on Midterm: final_hat = {a:.2f} + {b:.3f} * midterm")
print(f"Midterm on Final: midterm_hat = {a2:.2f} + {b2:.3f} * final")
```

```python
x_grid2 = np.linspace(x2.min(), x2.max(), 100)
y_hat_grid2 = a2 + b2 * x_grid2

plt.scatter(x2, y2, alpha=0.4)
plt.plot(x_grid2, y_hat_grid2, linewidth=2)
plt.xlabel("Final")
plt.ylabel("Midterm")
plt.title("Regression of Midterm on Final")
plt.show()
```

```python
# -----------------------------
# 2. Regression: y from x (Final | Midterm)
# -----------------------------
x_mean = x.mean()
y_mean = y.mean()
sx = x.std(ddof=1)
sy = y.std(ddof=1)
r = np.corrcoef(x, y)[0, 1]

b_yx = r * (sy / sx)
a_yx = y_mean - b_yx * x_mean

# -----------------------------
# 3. Regression: x from y (Midterm | Final)
# -----------------------------
b_xy = r * (sx / sy)
a_xy = x_mean - b_xy * y_mean

# Convert x|y line into y-form so both fit on same axes
x_grid = np.linspace(x.min(), x.max(), 200)
y_hat_y_from_x = a_yx + b_yx * x_grid               # y | x
y_hat_x_from_y = (x_grid - a_xy) / b_xy            # x | y converted

# -----------------------------
# 4. Plot both regression lines
# -----------------------------
plt.scatter(x, y, alpha=0.3)
plt.plot(x_grid, y_hat_y_from_x, linewidth=2, label="Predict Final from Midterm (y | x)")
plt.plot(x_grid, y_hat_x_from_y, linewidth=2, linestyle="--", label="Predict Midterm from Final (x | y)")

plt.xlabel("Final")
plt.ylabel("Midterm")
plt.title("Two Different Regression Lines on the Same Plot")
plt.legend()
plt.show()
```
