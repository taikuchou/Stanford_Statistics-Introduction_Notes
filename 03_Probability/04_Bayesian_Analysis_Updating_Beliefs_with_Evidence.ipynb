## Bayesian Analysis

### What is Bayesian Analysis?

- Bayesian analysis is a statistical method that updates beliefs (probabilities) when new evidence becomes available.

- It starts with a prior probability (initial belief), incorporates new evidence, and produces a posterior probability (updated belief).

The updating rule is Bayes’ Rule: **P(A∣B)= P(B∣A)⋅P(A)/P(B)** 


### Example: Email Spam Detection

- **Prior (before evidence)**: 20% of all emails are spam (P(spam)=0.2).
- **Evidence**: The word “money” appears in the email.
- **Likelihoods**:
    - P(money∣spam)=0.08
    - P(money∣ham)=0.01
- **Computation (using Bayes)**:
    - Denominator P(money)=0.08×0.2+0.01×0.8=0.024.
    - Posterior P(spam∣money)=0.08×0.2/0.024=0.667.

✅ Seeing “money” increases belief from 20% → **~67%** that the email is spam.

### Example: Medical Testing

- **Prior**: Disease prevalence is 1% (P(disease)=0.01).

- **Test performance**:
    - Sensitivity = 95% → P(positive∣disease)=0.95.
    - False positive rate = 2% → P(positive∣no disease)=0.02.

- **Computation**:
    - Denominator P(positive)=0.95×0.01+0.02×0.99=0.0095+0.0198=0.0293.
    - Posterior P(disease∣positive)=0.95×0.01/0.0293≈0.324.

✅ Even with a positive test, the chance of disease is **only ~32.4%**, not 95%.
This is because the disease is rare, so **false positives dominate**.

### 4. Why Bayesian Analysis Matters
- It teaches us to always consider the base rate (prior).
- Strong evidence (like a positive test) doesn’t guarantee certainty when the condition is rare.
- This reasoning is critical in spam filtering, diagnostics, fraud detection, and AI classifiers.

### ✅ Summary:

- Bayesian analysis = Prior belief + New evidence → Updated belief.
- It prevents misinterpretation by balancing how common something is (base rate) with how strong the evidence is (likelihoods).

## Warner’s Randomized Response Model

### The Problem

When surveys ask sensitive questions (e.g., “Have you ever cheated on an exam?”), respondents may lie because of **embarrassment, stigma, or fear of consequences**.
This leads to **biased results**—underreporting the true prevalence.

### Warner’s Randomized Response Technique (1965)

To encourage honesty while protecting privacy, Warner introduced randomization:

- Each respondent flips a coin **twice**.

- Based on the outcome, they answer **one of two questions**:

    1. **Sensitive Question**: “Have you ever cheated on an exam?”

    2. **Random Question**: “Did the coin land tails on the second toss?”
 
- If the first toss is tails, they answer Q1: “Have you ever cheated on an exam in college?”

- If the first toss is heads, they answer Q2: “Did you get tails on the second toss?”

The researcher doesn’t know which question the student answered, so **individual privacy is protected**.
But in aggregate, statistics can be used to estimate the true proportion.

### Why This Works

- A **“yes”** answer could mean:

    - The student cheated (from Q1), OR

    - The second coin landed tails (from Q2).

- Since the second coin toss is **random with known probability (50%)**, researchers can use probability theory to “filter out” the noise introduced by randomization.

### Example Calculation

**Suppose**:

- 50% of students answer Q1 (cheating), 50% answer Q2 (coin toss).

- Probability of “yes” from Q2 = 0.5.

If **47%** of all survey responses are “yes”: P(yes)=0.5⋅P(cheated)+0.5⋅0.5

**Solve for P(cheated)**:

- 0.47=0.5⋅P(cheated)+0.25
- 0.47−0.25=0.22=0.5⋅P(cheated)
- P(cheated)=0.44

✅ The estimated proportion of students who have cheated is **44%**, even though direct answers might have underreported it.

### Why It’s Important

- Ensures anonymity, encouraging honesty.
- Useful in surveys about **illegal, embarrassing, or sensitive behaviors** (e.g., drug use, tax evasion, sexual behavior).
- A brilliant application of **probability and conditional reasoning** in real-world research.

### Summary:
Warner’s Randomized Response Model introduces randomness to protect individuals while still allowing researchers to **uncover truthful group-level estimates**.

## Other examples

### Example 1: Drug Use Survey

- Directly asking “Have you ever used illegal drugs?” risks false answers.

- **Randomization**:

    - If the first coin toss = heads → answer the sensitive question (drug use).

    - If tails → answer: “Is your birthday in January, February, or March?”

- Since **~25%** of people will answer “yes” to the birthday question, we can subtract this noise to estimate the **true proportion of drug users**.

### Example 2: Tax Evasion

- Sensitive question: “Have you ever underreported income on your taxes?”
- **Randomization**:
    - Roll a die.
    - If 1 or 2 → answer “yes” no matter what.
    - If 3 or 4 → answer “no” no matter what.
    - If 5 or 6 → answer truthfully.
- Because we know the randomization structure (⅓ forced yes, ⅓ forced no, ⅓ truth), the aggregate results allow us to back out the **true fraction of tax evaders**.

### Example 3: Workplace Misconduct

- Question: “Have you ever harassed a colleague?” (very unlikely to be answered honestly).
- **Randomization with cards**:
    - Each employee draws a card from a shuffled deck.
    - If it’s a black card → answer truthfully.
    - If it’s a red card → automatically answer “yes.”
- The presence of forced “yes” answers makes it impossible to identify individuals but lets us calculate the **group-level misconduct prevalence**.

### Example 4: Political Opinion in Repressive Countries

- Sensitive question: “Do you support the opposition party?”
- **Randomization using dice**:
    - If you roll 1 → say “yes.”
    - If you roll 6 → say “no.”
    - Otherwise (2–5) → answer truthfully.
- Researchers can estimate **true support rates** while protecting individuals in environments where honest responses could be dangerous.

### Example 5: Academic Integrity (Exam Cheating)

- Toss two coins.
- If first coin is tails → answer the sensitive question (“Have you ever cheated?”).
- If first coin is heads → answer a random coin question (“Did second coin land tails?”).
- Aggregating answers, researchers estimate the true proportion of cheaters while preserving privacy.

### Key Idea in All Examples:

- Introduce random noise with known probability.
- This ensures individual anonymity, but the group’s true proportion can be estimated by adjusting for the noise.

```python
import random
import matplotlib.pyplot as plt
```

```python
# ---------- Warner's coin model ----------
# First coin flip chooses which question:
#   Tails -> answer sensitive truthfully (probability true_p of "yes")
#   Heads -> answer "did the second coin land tails?" (probability 0.5 of "yes")
# Estimator: p_hat = 2 * yes_rate - 0.5
def simulate_warner_coin(n, true_p, seed=42):
    rng = random.Random(seed)
    yes = 0
    for _ in range(n):
        first = rng.choice(("H", "T"))
        if first == "T":
            yes += (rng.random() < true_p)         # truthful on sensitive question
        else:
            second = rng.choice(("H", "T"))        # random noise question
            yes += (second == "T")
    y = yes / n
    p_hat = 2*y - 0.5
    return p_hat, y
```

```python
# ---------- Die forced yes/no model ----------
# Roll a die:
#   1-2: forced YES
#   3-4: forced NO
#   5-6: answer sensitive truthfully (probability true_p of "yes")
# Estimator: p_hat = 3 * yes_rate - 1
def simulate_die_forced(n, true_p, seed=42):
    rng = random.Random(seed)
    yes = 0
    for _ in range(n):
        d = rng.randint(1, 6)
        if d in (1, 2):
            yes += 1
        elif d in (3, 4):
            yes += 0
        else:
            yes += (rng.random() < true_p)
    y = yes / n
    p_hat = 3*y - 1
    return p_hat, y
```

```python
# ---------- Run experiments ----------
true_p = 0.44  # ground-truth cheating rate (can change)
sizes = [50, 100, 200, 500, 1000, 5000]

warner_estimates = []
die_estimates = []

for n in sizes:
    p_hat_w, _ = simulate_warner_coin(n, true_p, seed=42 + n)
    p_hat_d, _ = simulate_die_forced(n, true_p, seed=99 + n)
    warner_estimates.append(p_hat_w)
    die_estimates.append(p_hat_d)

print("True proportion (p):", true_p)
print("\nWarner coin model estimates:")
for n, est in zip(sizes, warner_estimates):
    print(f"  n={n:5d} -> p_hat={est:.4f}")

print("\nDie forced yes/no model estimates:")
for n, est in zip(sizes, die_estimates):
    print(f"  n={n:5d} -> p_hat={est:.4f}")
```

```python
# ---------- Plot convergence (one chart per model) ----------
plt.figure(figsize=(8,5))
plt.plot(sizes, warner_estimates, marker='o', label="Estimate")
plt.axhline(true_p, linestyle='--', label="True p")
plt.title("Warner Coin Model: Estimate vs Sample Size")
plt.xlabel("Sample size (n)")
plt.ylabel("Estimated p")
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,5))
plt.plot(sizes, die_estimates, marker='o', label="Estimate")
plt.axhline(true_p, linestyle='--', label="True p")
plt.title("Die Forced Yes/No Model: Estimate vs Sample Size")
plt.xlabel("Sample size (n)")
plt.ylabel("Estimated p")
plt.legend()
plt.tight_layout()
plt.show()
```

### Bias
- Both designs are **approximately unbiased** — their average estimate is close to the true value (here, p=0.44).
- As **sample size (n)** increases, the bias shrinks further toward zero.

### Variance
- **Warner’s coin model** has **much lower variance**. The randomness it introduces (coin flip + second coin toss) is relatively mild.
- **Die forced yes/no** has **higher variance** because 2/3 of responses are artificially forced (yes/no), leaving only 1/3 actually informative. That adds a lot of noise.
  
### RMSE (Root Mean Squared Error)
- RMSE combines both **bias and variance** → it’s a measure of overall accuracy.
- **Warner’s model consistently has lower RMSE** across all sample sizes.
- **Die model RMSE is larger**, meaning you need a much bigger sample to reach the same accuracy Warner’s method gets with fewer participants.

### Practical Implications
- If you want **higher efficiency per respondent**, Warner’s coin model is better: less noisy, faster convergence.
- The die forced model may be easier to explain in practice, but it “wastes” information by forcing answers, making it statistically less efficient.
- In real surveys, Warner’s model would typically be preferred unless operational constraints make the die easier to use.

### ✅ Bottom line:
- Both protect privacy and allow truthful prevalence estimation.
- **Warner’s model = lower variance, lower RMSE, more efficient**.
- **Die forced yes/no = noisier, needs larger samples to stabilize**.
