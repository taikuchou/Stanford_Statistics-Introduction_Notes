## Solving problems by total enumeration (law of total probability)

**1. Conditional Probability Refresher**

- Conditional probability tells us the likelihood of event B happening, given that event A has already happened.
 **P(B∣A)=P(A and B)/P(A)**

- Rearranging gives the **general multiplication rule**: **P(A and B)=P(A)×P(B∣A)**

- If A and B are independent, then **P(B∣A)=P(B)**, so it simplifies to the usual multiplication rule.

**2. Law of Total Probability**

When we want the probability of some event B, and the sample space is split into cases (like “spam” vs “ham”), we add up the contributions from each case:

P(B)=P(B∣A1)P(A1)+P(B∣A2)P(A2)+…

This is called the **law of total probability**, or “solving by total enumeration.”


**3. Email Example**

We want the probability that a random email contains the word **“money”**

- 20% of emails are **spam** → P(spam)=0.2
- 80% of emails are **ham** → P(ham)=0.8
- In spam emails, “money” appears 8% of the time → P(money∣spam)=0.08
- In ham emails, “money” appears 1% of the time → P(money∣ham)=0.01

Now apply the law of total probability:

**P(money)**

=P(money∣spam)⋅P(spam)+P(money∣ham)⋅P(ham)

=0.08×0.2+0.01×0.8

=0.016+0.008

=0.024


```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# Probabilities
p_spam = 0.2
p_ham = 0.8
p_money_given_spam = 0.08
p_money_given_ham = 0.01

# Joint probabilities
p_money_and_spam = p_spam * p_money_given_spam   # 0.016
p_nomoney_and_spam = p_spam * (1 - p_money_given_spam) # 0.184
p_money_and_ham = p_ham * p_money_given_ham     # 0.008
p_nomoney_and_ham = p_ham * (1 - p_money_given_ham)   # 0.792
```

```python
# Build graph
G = nx.DiGraph()

# Nodes
G.add_node("Email")
G.add_node("Spam (0.2)")
G.add_node("Ham (0.8)")
G.add_node(f"Money | Spam ({p_money_and_spam:.3f})")
G.add_node(f"No Money | Spam ({p_nomoney_and_spam:.3f})")
G.add_node(f"Money | Ham ({p_money_and_ham:.3f})")
G.add_node(f"No Money | Ham ({p_nomoney_and_ham:.3f})")

# Edges with conditional probabilities
G.add_edge("Email", "Spam (0.2)", weight=p_spam)
G.add_edge("Email", "Ham (0.8)", weight=p_ham)

G.add_edge("Spam (0.2)", f"Money | Spam ({p_money_and_spam:.3f})", weight=p_money_given_spam)
G.add_edge("Spam (0.2)", f"No Money | Spam ({p_nomoney_and_spam:.3f})", weight=1-p_money_given_spam)

G.add_edge("Ham (0.8)", f"Money | Ham ({p_money_and_ham:.3f})", weight=p_money_given_ham)
G.add_edge("Ham (0.8)", f"No Money | Ham ({p_nomoney_and_ham:.3f})", weight=1-p_money_given_ham)

# Layout
pos = {
    "Email": (0, 2),
    "Spam (0.2)": (-2, 1),
    "Ham (0.8)": (2, 1),
    f"Money | Spam ({p_money_and_spam:.3f})": (-3, 0),
    f"No Money | Spam ({p_nomoney_and_spam:.3f})": (-1, 0),
    f"Money | Ham ({p_money_and_ham:.3f})": (1, 0),
    f"No Money | Ham ({p_nomoney_and_ham:.3f})": (3, 0),
}

# Draw graph
plt.figure(figsize=(11,6))
nx.draw(G, pos, with_labels=True, node_size=3400, node_color="lightyellow",
        font_size=9, font_weight="bold", arrows=True)

# Edge labels (conditional probs)
edge_labels = {(u, v): f"{d['weight']:.2f}" for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=9)

plt.title("Tree Diagram with Joint Probabilities (Law of Total Probability)", fontsize=12)
plt.axis("off")
plt.show()

# Final result
print("P(money) =", p_money_and_spam + p_money_and_ham)
```

### 1. What Bayes’ Rule Does

Bayes’ Rule lets us invert conditional probabilities.
Instead of asking “What is the probability that ‘money’ appears in spam?” we ask “Given that ‘money’ appears, what is the probability the email is spam?”

Mathematically:

P(spam∣money)= P(money∣spam)⋅P(spam)/P(money)

### 2. Applying It to Spam Emails

From the problem:

- P(spam)=0.20 (20% of emails are spam)
- P(ham)=0.80
- P(money∣spam)=0.08 (8% of spam emails have “money”)
- P(money∣ham)=0.01

First, use the **law of total probability** to find 

P(money)

= P(money∣spam)⋅P(spam)+P(money∣ham)⋅P(ham)

= 0.08×0.2+0.01×0.8

= 0.016+0.008

= 0.024

So, 2.4% of all emails contain “money.”

### 3. Compute Posterior Probability

Now apply Bayes’ Rule:

P(spam∣money)= 0.08×0.20/0.024 = 0.016/0.024 ≈ 0.667

So the probability that an email is spam **given it contains “money”** is about 67%.

## Bayes' Theorem + Total Probability + Simulation + Simple Confusion Table
Scenario: spam/ham emails and the word "money"

```python
import random
import matplotlib.pyplot as plt
```

```python
# -----------------------------
# Given probabilities
# -----------------------------
p_spam = 0.20
p_ham = 1.0 - p_spam

p_money_given_spam = 0.08
p_money_given_ham  = 0.01
```

```python
# -----------------------------
# Law of Total Probability
# -----------------------------
p_money = p_money_given_spam * p_spam + p_money_given_ham * p_ham
p_no_money = 1.0 - p_money
p_money,p_no_money
```

```python
# -----------------------------
# Bayes' Theorem (theoretical)
# -----------------------------
# P(spam | money)
p_spam_given_money = p_money_given_spam * p_spam / p_money
# P(ham | money)
p_ham_given_money  = p_money_given_ham  * p_ham  / p_money
p_spam_given_money, p_ham_given_money
```

```python
# Also for "no money"
# P(spam | no money) and P(ham | no money)
p_no_money_given_spam = 1.0 - p_money_given_spam
p_no_money_given_ham  = 1.0 - p_money_given_ham

p_spam_given_no_money = p_no_money_given_spam * p_spam / p_no_money
p_ham_given_no_money  = p_no_money_given_ham  * p_ham  / p_no_money
p_spam_given_no_money,p_ham_given_no_money
```

```python
# -----------------------------
# Simulation to verify
# -----------------------------
trials = 500_000
spam_and_money = 0
ham_and_money  = 0
spam_and_no_money = 0
ham_and_no_money  = 0

for _ in range(trials):
    is_spam = random.random() < p_spam
    if is_spam:
        has_money = random.random() < p_money_given_spam
        if has_money:
            spam_and_money += 1
        else:
            spam_and_no_money += 1
    else:
        has_money = random.random() < p_money_given_ham
        if has_money:
            ham_and_money += 1
        else:
            ham_and_no_money += 1
money_total    = spam_and_money + ham_and_money
no_money_total = spam_and_no_money + ham_and_no_money
money_total,no_money_total
```

```python
p_spam_given_money_sim    = spam_and_money / money_total
p_ham_given_money_sim     = ham_and_money  / money_total
p_spam_given_no_money_sim = spam_and_no_money / no_money_total
p_ham_given_no_money_sim  = ham_and_no_money  / no_money_total

p_spam_given_money_sim,p_ham_given_money_sim,p_spam_given_no_money_sim,p_ham_given_no_money_sim
```

```python
# -----------------------------
# Print results
# -----------------------------
print("=== Theoretical ===")
print(f"P(money)                 = {p_money:.6f}")
print(f"P(spam | money)          = {p_spam_given_money:.6f}")
print(f"P(ham  | money)          = {p_ham_given_money:.6f}")
print(f"P(spam | no money)       = {p_spam_given_no_money:.6f}")
print(f"P(ham  | no money)       = {p_ham_given_no_money:.6f}")

print("\n=== Simulation ===")
print(f"P(spam | money)   ~ {p_spam_given_money_sim:.6f}")
print(f"P(ham  | money)   ~ {p_ham_given_money_sim:.6f}")
print(f"P(spam | no money)~ {p_spam_given_no_money_sim:.6f}")
print(f"P(ham  | no money)~ {p_ham_given_no_money_sim:.6f}")

print("\n=== Approx. confusion-style counts (simulation) ===")
print("                money        no_money")
print(f"spam       {spam_and_money:10d}   {spam_and_no_money:10d}")
print(f"ham        {ham_and_money:10d}   {ham_and_no_money:10d}")
print(f"TOTAL      {money_total:10d}   {no_money_total:10d}")
```

```python
# -----------------------------
# Plot comparison (posteriors given 'money')
# -----------------------------
labels = ["P(spam | money)", "P(ham | money)",
          "P(spam | no money)", "P(ham | no money)"]
theoretical_vals = [p_spam_given_money, p_ham_given_money,
                    p_spam_given_no_money, p_ham_given_no_money]
simulated_vals   = [p_spam_given_money_sim, p_ham_given_money_sim,
                    p_spam_given_no_money_sim, p_ham_given_no_money_sim]

x = range(len(labels))
plt.figure(figsize=(10,6))
plt.bar(x, theoretical_vals, width=0.4, label="Theoretical")
plt.bar([i + 0.4 for i in x], simulated_vals, width=0.4, label="Simulated")
plt.xticks([i + 0.2 for i in x], labels, rotation=15)
plt.ylabel("Probability")
plt.title("Bayes' Theorem Posteriors: money vs. no money")
plt.legend()
plt.tight_layout()
plt.show()
```
