## Example 1: Medical Test Accuracy

Suppose a disease affects 1% of the population. 

A medical test is:

- 99% accurate if someone has the disease (true positive rate)

- 95% accurate if someone does not have the disease (true negative rate)

**Question**: If a patient tests positive, what’s the probability they actually have the disease?

### Step 1. Define probabilities

- P(disease)=0.01
- P(no disease)=0.99
- P(positive∣disease)=0.99
- P(positive∣no disease)=0.05 (false positive rate = 1 - 0.95)

### Step 2. Total probability of a positive test
P(positive)

= P(positive∣disease)P(disease)+P(positive∣no disease)P(no disease)

= (0.99×0.01)+(0.05×0.99)

= 0.0099+0.0495

= 0.0594

### Step 3. Apply Bayes’ Rule

P(disease∣positive)

= P(positive∣disease)P(disease) / P(positive)

= 0.0099 / 0.0594

≈ **0.167**

Even with a positive test, the chance of actually having the disease is only **~16.7%** because the disease is rare.

## Example 2: Factory Defective Items

Two factories produce widgets:

- Factory A: 40% of total production, with 1% defect rate

- Factory B: 60% of total production, with 2% defect rate

**Question**: If a widget is defective, what’s the probability it came from Factory B?


### Step 1. Define probabilities
- P(A)=0.4, P(B)=0.6
- P(defect∣A)=0.01
- P(defect∣B)=0.02

### Step 2. Total probability of defect

**P(defect)**

=(0.01×0.4)+(0.02×0.6)

=0.004+0.012

=0.016

### Step 3. Apply Bayes’ Rule

**P(B∣defect)** 

=P(defect∣B)P(B) / P(defect)

=0.02×0.6 / 0.016

=0.012 / 0.016

=0.75

## Example 3: Weather Forecasting

- If it doesn’t rain, the probability you still see dark clouds is P(clouds∣no rain)=0.2.

**Question**: If you see dark clouds, what’s the probability it will rain?
