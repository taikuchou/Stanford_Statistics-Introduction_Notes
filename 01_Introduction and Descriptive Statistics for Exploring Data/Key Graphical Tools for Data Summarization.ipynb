### Graphical summaries help represent data visually, and the choice depends on whether the data is categorical or numerical.

**Qualitative Data (Categorical)**
Pie chart: Shows categories as slices of a circle, highlighting proportions of the whole but making direct comparisons harder.

Dot plot: Uses dots along a line to represent percentages, allowing easier side-by-side comparisons but less intuitive for overall proportions.

**Quantitative Data (Numerical)**
Bar graph: Uses bars to represent counts or measures, good for showing numerical values along a number line.

Histogram: Similar to a bar graph but bars may vary in width. Bar area represents proportions of the data, total area equals 100%. Bar height shows density, making it useful for understanding distribution and percentages within ranges.

**In short, pie charts and dot plots are best for categorical data, while bar graphs and histograms are most effective for numerical data. Pie charts emphasize proportions, dot plots ease**

### matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sample data
data = np.random.randn(1000)  # for histogram
categories = ['A', 'B', 'C', 'D']
values = [15, 30, 45, 10]
```

```python
# --- Pie Chart ---
plt.figure(figsize=(6, 6))
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart Example")
plt.show()
```

```python
# --- Dot Plot ---
plt.figure(figsize=(6, 4))
x = np.arange(len(categories))
y = values
plt.plot(x, y, 'o', markersize=10)
plt.xticks(x, categories)
plt.title("Dot Plot Example")
plt.show()
```

```python
# --- Bar Graph ---
plt.figure(figsize=(6, 4))
plt.bar(categories, values, color='skyblue')
plt.title("Bar Graph Example")
plt.show()
```

```python
# --- Histogram ---
plt.figure(figsize=(6, 4))
plt.hist(data, bins=30, color='lightgreen', edgecolor='black')
plt.title("Histogram Example")
plt.show()
```

### seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Sample data
data = np.random.randn(1000)  # for histogram
categories = ['A', 'B', 'C', 'D']
values = [15, 30, 45, 10]
df = pd.DataFrame({"Category": categories, "Value": values})
```

```python
# --- Pie Chart (matplotlib, since Seaborn has no native pie) ---
plt.figure(figsize=(6, 6))
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart Example (Matplotlib)")
plt.show()
```

```python
# --- Dot Plot ---
plt.figure(figsize=(6, 4))
sns.scatterplot(x="Category", y="Value", data=df, s=100, color="red")
plt.title("Dot Plot Example (Seaborn)")
plt.show()
```

```python
# --- Bar Graph ---
plt.figure(figsize=(6, 4))
sns.barplot(x="Category", y="Value", data=df, palette="Blues_d")
plt.title("Bar Graph Example (Seaborn)")
plt.show()
```

```python
# --- Histogram ---
plt.figure(figsize=(6, 4))
sns.histplot(data, bins=30, kde=True, color="green")
plt.title("Histogram Example (Seaborn)")
plt.show()
```

### 2x2 grid of charts

```python
# Create 2x2 grid
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# --- Pie Chart ---
axes[0,0].pie(values, labels=categories, autopct='%1.1f%%', startangle=90)
axes[0,0].set_title("Pie Chart")

# --- Dot Plot ---
sns.scatterplot(x="Category", y="Value", data=df, s=100, color="red", ax=axes[0,1])
axes[0,1].set_title("Dot Plot")

# --- Bar Graph ---
sns.barplot(x="Category", y="Value", data=df, palette="Blues_d", ax=axes[1,0])
axes[1,0].set_title("Bar Graph")

# --- Histogram ---
sns.histplot(data, bins=30, kde=True, color="green", ax=axes[1,1])
axes[1,1].set_title("Histogram")

plt.tight_layout()
plt.show()
```

### Box Plot (Box and Whisker)
Represents minimum, first quartile, median, third quartile, and maximum values.

Useful for comparing distributions between groups side by side without choosing intervals.

Compact and good for summary comparisons, but shows less detail than a histogram.

### Scatter Plot
Plots paired data points on a grid, showing one variable on each axis.

Reveals trends or relationships between two variables, including linear and non-linear patterns.

Essential for bivariate data, highlighting correlations and outliers.

**In summary, box plots efficiently compare multiple distributions, while scatter plots visualize connections and trends between two variables. Both provide fast, intuitive data summaries for exploratory analysis.**



```python
# --- Sample Data ---
np.random.seed(42)
data = np.random.randn(100, 4)  # 100 rows, 4 categories
categories = ["A", "B", "C", "D"]
df = pd.DataFrame(data, columns=categories)
```

```python
# Another dataset for scatter plot
x = np.random.randn(100)
y = 2 * x + np.random.randn(100) * 0.5  # Linear relationship with noise
```

```python
def save_to_file(output_folder,file_name,show=True):
    file_path = os.path.join(output_folder, file_name)
    plt.savefig(file_path, dpi=300)
    if not show:
        plt.close()
    print(f" Saved graph to: {file_path}")
```

```python
# --- Box Plot (Box and Whisker) ---
plt.figure(figsize=(6,4))
sns.boxplot(data=df, palette="Set2")
plt.title("Box Plot (Box-and-Whisker)")
plt.xlabel("Category")
plt.ylabel("Values")
save_to_file(output_folder,"box_plot.png")
plt.show()
```

```python
# --- Scatter Plot ---
plt.figure(figsize=(6,4))
plt.scatter(x, y, alpha=0.7, c="blue", edgecolors="k")
plt.title("Scatter Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values")
save_to_file(output_folder,"scatter_plot.png")
plt.show()
```

### Iris flower dataset  (150 samples, 3 species: Setosa, Versicolor, Virginica).

```python
# Load the classic Iris dataset
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
species = iris['species'].unique()
list(species)
```

```python
# --- Pie Chart: Species distribution ---
species_counts = iris['species'].value_counts()
plt.figure(figsize=(5,5))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart: Iris Species Distribution")
plt.show()
```

**Species Distribution (Pie Chart)**

- Each of the three species (Setosa, Versicolor, Virginica) has 50 samples (33.3% each).**

- This balanced dataset ensures fair comparisons across species.**

```python
# --- Dot Plot: Sepal length per species ---
plt.figure(figsize=(6,4))
sns.stripplot(x="species", y="sepal_length", data=iris, jitter=True, size=5)
plt.title("Dot Plot: Sepal Length by Species")
plt.show()
```

**Sepal Length (Dot Plot)**

- Setosa has consistently shorter sepal lengths.

- Versicolor and Virginica overlap, but Virginica generally has longer sepals.

```python
# --- Bar Graph: Average petal length per species ---
plt.figure(figsize=(6,4))
sns.barplot(x="species", y="petal_length", data=iris, ci=None, palette="Blues_d")
plt.title("Bar Graph: Average Petal Length by Species")
plt.show()
```

**Petal Length (Bar Graph)**

- Petal length is a key differentiator:

  - Setosa: short petals

  - Versicolor: medium length

  - Virginica: longest petals


```python
# --- Histogram: Sepal width distribution ---
plt.figure(figsize=(6,4))
sns.histplot(iris['sepal_width'], bins=20, kde=True, color="green")
plt.title("Histogram: Sepal Width Distribution")
plt.show()
```

```python
# Create 3x1 grid
species_list = iris['species'].unique()
# Pick a color palette with the same length as species
palette = sns.color_palette("Set2", len(species_list))
# Create 1x3 grid (3 columns, 1 row)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot histograms for each species
for i, s in enumerate(species_list):
    sns.histplot(iris[iris['species'] == s]['sepal_width'], bins=20, kde=True, ax=axes[i],color=palette[i])
    axes[i].set_title(f"{s.upper()} ")

plt.tight_layout()
plt.show()
```

**Sepal Width (Histogram)**

- Sepal width distribution is **more varied**.

- Setosa tends to have **wider sepals**, while Virginica and Versicolor overlap more.

- Not as strong a distinguishing factor compared to petal length.

```python
# --- Box Plot: Petal length across species ---
plt.figure(figsize=(6,4))
sns.boxplot(x="species", y="petal_length", data=iris, palette="Set2")
plt.title("Box Plot: Petal Length by Species")
plt.show()
```

**Petal Length (Box Plot)**
- Box Plot confirms: clear separation, with almost **no overlap** between Setosa and others  strong feature for classification.

```python
# --- Scatter Plot: Sepal length vs Sepal width ---
plt.figure(figsize=(6,4))
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=iris, s=70, palette="Set1")
plt.title("Scatter Plot: Sepal Length vs Sepal Width")
plt.show()
```

**Sepal Length vs Width (Scatter Plot)**

- Setosa forms a distinct cluster (short sepal length, wide sepal width).

- Versicolor and Virginica overlap, but Virginica leans toward larger sepal length.

- Scatter plot shows why machine learning classification on Iris often uses petal features over sepal features.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load Iris dataset
iris = sns.load_dataset("iris")

# Count for Pie Chart
species_counts = iris['species'].value_counts()

# Create 3x2 grid
fig, axes = plt.subplots(3, 2, figsize=(14, 16))

# --- Pie Chart: Species distribution ---
axes[0,0].pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=90)
axes[0,0].set_title("Pie Chart: Species Distribution")

# --- Dot Plot: Sepal length per species ---
sns.stripplot(x="species", y="sepal_length", data=iris, jitter=True, size=5, ax=axes[0,1])
axes[0,1].set_title("Dot Plot: Sepal Length by Species")

# --- Bar Graph: Average petal length per species ---
sns.barplot(x="species", y="petal_length", data=iris, ci=None, palette="Blues_d", ax=axes[1,0])
axes[1,0].set_title("Bar Graph: Avg Petal Length by Species")

# --- Histogram: Sepal width distribution ---
sns.histplot(iris['sepal_width'], bins=20, kde=True, color="green", ax=axes[1,1])
axes[1,1].set_title("Histogram: Sepal Width Distribution")

# --- Box Plot: Petal length across species ---
sns.boxplot(x="species", y="petal_length", data=iris, palette="Set2", ax=axes[2,0])
axes[2,0].set_title("Box Plot: Petal Length by Species")

# --- Scatter Plot: Sepal length vs Sepal width ---
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=iris, s=70, palette="Set1", ax=axes[2,1])
axes[2,1].set_title("Scatter Plot: Sepal Length vs Sepal Width")

# Adjust layout
plt.tight_layout()
plt.show()
```
