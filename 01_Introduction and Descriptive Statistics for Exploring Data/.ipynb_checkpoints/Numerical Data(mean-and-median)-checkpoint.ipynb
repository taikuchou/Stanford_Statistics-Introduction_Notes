### Mean-and-median
Numerical summary measures help describe large datasets with just a few numbers. The most common are the **mean, median, and percentiles** (including quartiles), and which one you use depends on the shape of the data distribution.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Generate some skewed data
np.random.seed(42)
data = np.random.exponential(scale=2.0, size=1000)  # Skewed distribution
```

```python
# --- Numerical Summary Measures ---
mean_val = np.mean(data)
median_val = np.median(data)
q1 = np.percentile(data, 25)   # 25th percentile
q3 = np.percentile(data, 75)   # 75th percentile
iqr = q3 - q1                  # Interquartile Range

print("ðŸ“Š Numerical Summary Measures")
print(f"Mean: {mean_val:.2f}")
print(f"Median: {median_val:.2f}")
print(f"Q1 (25th percentile): {q1:.2f}")
print(f"Q3 (75th percentile): {q3:.2f}")
print(f"IQR (Q3 - Q1): {iqr:.2f}")
```

```python
# --- Visualization ---
plt.figure(figsize=(10,5))
sns.histplot(data, bins=30, kde=True, color="lightblue", edgecolor="black")

# Add vertical lines for mean, median, and quartiles
plt.axvline(mean_val, color="red", linestyle="--", linewidth=2, label=f"Mean = {mean_val:.2f}")
plt.axvline(median_val, color="green", linestyle="-", linewidth=2, label=f"Median = {median_val:.2f}")
plt.axvline(q1, color="orange", linestyle="--", label=f"Q1 = {q1:.2f}")
plt.axvline(q3, color="purple", linestyle="--", label=f"Q3 = {q3:.2f}")

plt.title("Mean, Median, and Quartiles on Skewed Data")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate skewed data
np.random.seed(42)
data = np.random.exponential(scale=2.0, size=1000)

# Convert to DataFrame for Seaborn
df = pd.DataFrame({"Value": data})

# --- Box Plot ---
plt.figure(figsize=(8,5))
sns.boxplot(x="Value", data=df, color="skyblue")

plt.title("Box Plot: Median, Quartiles, and Outliers")
plt.xlabel("Value")
plt.show()
```

## Example: Housing Prices (Synthetic Data)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Simulated house price dataset (in $1000s)
np.random.seed(42)
low_prices = np.random.normal(200, 30, 500)     # majority of homes
high_prices = np.random.normal(1000, 150, 20)   # luxury homes (outliers)
prices = np.concatenate([low_prices, high_prices])

df = pd.DataFrame({"House Price ($1000s)": prices})
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
print("Mean:\n", df.mean(numeric_only=True))
print('\n')
print("Median:\n", df.median(numeric_only=True))
print('\n')
print("Mode:\n", df.mode().iloc[0])
print('\n')
print("Variance:\n", df.var(numeric_only=True))
print('\n')
print("Standard Deviation:\n", df.std(numeric_only=True))
print('\n')
print("Min:\n", df.min(numeric_only=True))
print('\n')
print("Max:\n", df.max(numeric_only=True))
print('\n')
print("Quantiles:\n", df.quantile([0.25, 0.5, 0.75]))
print('\n')
```

```python
# --- Summary Statistics ---
# mean_val = np.mean(prices)
# median_val = np.median(prices)

# print("ðŸ“Š Summary Statistics")
# print(f"Mean House Price: {mean_val:.2f}k")
# print(f"Median House Price: {median_val:.2f}k")
# --- Statistics using NumPy APIs ---
arr = prices
print("ðŸ“Š NumPy Summary Statistics")
print("Count:", arr.size)
print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Variance:", np.var(arr, ddof=1))   # ddof=1 -> sample variance
print("Standard Deviation:", np.std(arr, ddof=1))
print("Min:", np.min(arr))
print("Max:", np.max(arr))
print("25th Percentile (Q1):", np.percentile(arr, 25))
print("50th Percentile (Median):", np.percentile(arr, 50))
print("75th Percentile (Q3):", np.percentile(arr, 75))
```

```python
# --- Visualization ---
plt.figure(figsize=(10,5))
sns.histplot(prices, bins=40, kde=True, color="skyblue", edgecolor="black")

plt.axvline(mean_val, color="red", linestyle="--", linewidth=2, label=f"Mean = {mean_val:.2f}k")
plt.axvline(median_val, color="green", linestyle="-", linewidth=2, label=f"Median = {median_val:.2f}k")

plt.title("House Price Distribution: Mean vs Median")
plt.xlabel("House Price ($1000s)")
plt.ylabel("Count")
plt.legend()
plt.show()
```
